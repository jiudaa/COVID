```python
import pandas as pd
data = pd.read_csv('./covid-policy-tracker-master/data/OxCGRT_latest.csv')
```

```python
data.columns
```

```python
data.CountryCode.unique()
```

```python
import matplotlib.pyplot as plt
df = data.sort_values('Date', ascending=True)
df['Date'] = pd.to_datetime(df['Date'], format = '%Y%m%d')
# owid.tail(20)

df_filtered = df[df.CountryCode=='NLD']
with plt.style.context('fivethirtyeight'):
    df_filtered = df[df.CountryCode=='NLD']
    df_filtered.plot(x="Date", y=["GovernmentResponseIndex", "ContainmentHealthIndex", "EconomicSupportIndex"])
    plt.xticks(rotation=90)
    plt.title('Plotting new cases, total cases and total cases per million in the Netherlands\n', size=15);
    plt.show()
```
